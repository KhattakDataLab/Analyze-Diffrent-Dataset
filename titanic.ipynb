## a bar chart showing the number of passengers by class (Pclass).

```python
# counts 
pclass_counts=df['Pclass'].value_counts()
```

```python
pclass_counts
```

```python
plt.bar(pclass_counts.index, pclass_counts.values, color='aqua')
# Adding title and labels
plt.title('Number of Passengers by Class')
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.xticks(pclass_counts.index)

# Show the plot
plt.show()
```

### House pricing Data Set

```python
df1=pd.read_csv('house.csv')
```

```python
df1
```

```python
# Randomly sample 200 rows from the dataset
df_sample = df1.sample(n=200, random_state=1)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_sample['total_sqft'], df_sample['price'], color='hotpink')

# Adding title and labels
plt.title('Total Square Feet vs. Price')
plt.xlabel('Total Squar')
plt.ylabel('Price')

# Show the plot
plt.show()
```

# Explore Heart Diseasse Data Set


```python
# load the data set 
df2=pd.read_csv('heart.csv')
```

```python
df2
```

```python
df2.columns
```

```python
# Plot histogram for 'age'
plt.figure(figsize=(10, 6))
plt.hist(df2['age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df2['sex'].dropna(), bins=30, color='aqua', edgecolor='red')
plt.title('Sex Distribution of Patients')
plt.xlabel('sex')
plt.ylabel('Number of Patients')
plt.show()
```

## iris DataSet Exploration 

```python
df3=pd.read_csv('iris.csv')
```

```python
df3
```

```python
# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_length', data=df3, palette='Set2')

# Adding title and labels
plt.title('Distribution of Petal Lengths for Each Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Show the plot
plt.show()
```
